```python
import numpy as np
```

```python
## создание массивов
list_data = [1, 2, 3, 4, 5]
arr_from_list = np.array(list_data)
print(f"Массив из списка: {arr_from_list}")
arr_zeros = np.zeros((2, 3))
print(f"Массив нулей (2x3):\n{arr_zeros}")
arr_ones = np.ones((1, 4))
print(f"Массив единиц (1x4): {arr_ones}")
arr_range = np.arange(0, 10, 2)
print(f"Массив из диапазона (arange): {arr_range}")
print("\n")
```

```python
## атрибуты 
print("--- 3. Атрибуты массива (arr_from_list) ---")
print(f"Массив: {arr_from_list}")
print(f"Форма (shape): {arr_from_list.shape}")
print(f"Тип данных (dtype): {arr_from_list.dtype}")
print("\n")
```

```python
## срезы и индексирование
print("--- 4. Индексирование и срезы (arr_range) ---")
print(f"Исходный массив: {arr_range}")
element = arr_range[2]
print(f"Третий элемент (индекс 2): {element}")
subarray_slice = arr_range[0:3]
print(f"Подмассив (срез 0:3): {subarray_slice}")
print("\n")
```

```python
## статистика
print("--- 5. Векторные операции и статистика ---")
arr1 = np.array([10, 20, 30])
arr2 = np.array([1, 2, 3])
print(f"Массив 1: {arr1}")
print(f"Массив 2: {arr2}")
arr_sum = arr1 + arr2
print(f"Поэлементное сложение: {arr_sum}")
arr_prod = arr1 * arr2
print(f"Поэлементное умножение: {arr_prod}")
arr1_mean = np.mean(arr1)
print(f"Среднее значение arr1: {arr1_mean}")
arr2_sum = np.sum(arr2)
print(f"Сумма значений arr2: {arr2_sum}")
```
